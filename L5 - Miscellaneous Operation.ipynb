Apply user defined Functions

```python
# Define function which converts execution time from second to milli_seconds
def testfunc(x):
    model_name = "_".join(x.split())
    model_name = model_name.lower()
    return model_name
```

```python
#convert model name to lower case and space by undersocre
df["model_name"].apply(testfunc)
```

```python
df
```

```python
# using map() instead of apply
df["model_name"].map(testfunc)
```

```python
df
```

```python
# convert execution time from second to milli_second - using apply
df["execution_time"].apply(lambda x : x*1000)
```

```python
def change_time(x):
    return x*1000
```

```python
# COnvert execution time from second to milli-second - using map
df["execution_time"].map(change_time)
```

```python
df
```

```python
# model_score = 0.8"f1_score +0.1*(1/execution_time)+ 0.1*(1/model_size)
df["model_score"]= (
    0.8*df["f1_score"]+
    0.1*(1/df["execution_time"])+
    0.1*(1/df["model_size"])
    
)
```

```python
df
```

```python
# Even through Random Forest model have high f1 score than decision tree , the overall score for Random Forest is less than Decison Tres
```

Benefits of DateTimeIndex over regular columns

```python
#sample data
data = {
    "Date":pd.date_range("2025-08-18", periods=5 , freq="D"),
    "Value":[10,20,30,40,50]

}
df_col=pd.DataFrame(data)
#DataFrame with DateTimeIndex
df_index= df_col.drop( columns="Date")
df_index.index = data["Date"]

print("DataFrame with datetime columns :\n", df_col)
print("\nDataFrame with DatetimeIndex:\n", df_index)
```

```python
# flitering a specfic data
# using date time column
print(df_col[df_col["Date"] == "2025-8-20"])
```

```python
#using DatetimIndex
df_index.loc["2025-8-20"]
```

```python
# slicing by date range 
# using datetime columns
df_col[(df_col["Date"] >= "2025-08-19") & (df_col["Date"]<= "2025-08-21")]
```

```python
# using datetimeIndex
df_index["2025-08-19":"2025-08-21"]
```
