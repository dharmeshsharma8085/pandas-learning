```python
# import duckdb
import duckdb
```

```python
# Create a database conection 
con = duckdb.connect("Onlinestrore.duckdb")
```

```python
# Q1 Creating a customer table
con.execute(
    """
    CREATE TABLE Customers (
    customer_id INT PRIMARY KEY ,
    customer_name VARCHAR(50),
    city VARCHAR(50),
    age INT);
    """  
)
```

```python
#Q2 Insert customer in to Customers table
con.execute(
    """
    INSERT INTO Customers (customer_id , customer_name , city , age) VALUES
    (1  ,'RAHUL' ,'DELHI' , 28),
    (2  ,'AMIT' ,'LUCKNOW' , 29),
    (3  ,'LUCKY' ,'AMRITSAR' , 26),
    (4  ,'RAMESH' ,'CHENNAI' , 34),
    (5  ,'SANKET' ,'PUNE' , 20);
    """  
) 
```

```python
#Q3 Cretae order table
con.execute(
    """
    CREATE TABLE Orders(
        order_id INT PRIMARY KEY ,
        customer_id INT,
        product VARCHAR(50),
        price INT,
        order_date DATE,
        FOREIGN KEY ( customer_id) REFERENCES Customers(customer_id));
    """    
)
```

```python
# Q4 add order in orders table
con.execute(
    """
    INSERT INTO Orders (order_id , customer_id , product , price , order_date)VALUES
    (101 , 1 , 'Shoes' , 1200 , '2025-08-01'),
    (102 , 1 , 'Bag' , 800 , '2025-08-05'),
    (103 , 1 , 'Laptop' , 60000 , '2025-08-03'),
    (104 , 1 , 'Shoes' , 1000 , '2025-08-04'),
    (105 , 1 , 'Mobile' , 20000 , '2025-08-06'),
    (106 , 1 , 'Shoes' , 1500 , '2025-08-02'),
    (107 , 1 , 'Bag' , 700 , '2025-08-07'),
    (108 , 1 , 'Headphones' , 2000 , '2025-08-08'),
    (109 , 1 , 'Shoes' , 1100 , '2025-08-09'),
    (110 , 1 , 'Laptop' , 58000 , '2025-08-10');

    """
    
)
```

```python
# IF we want to see complete table
con.execute("SELECT*FROM Orders").fetchall()
```

```python
# Q5 show the customer name and city
con.execute("SELECT  customer_name , city  FROM Customers;").fetchall()
```

```python
# Q6 show only unique prdocuts from the order table
con.execute("SELECT DISTINCT product FROM Orders;").fetchall()
```

```python
#Q7 Filters rows in orders where price is gretaer then a 1000
con.execute("SELECT*FROM Orders WHERE price>1000;").fetchall()
```

```python
#Q8 Retrieve customers in customer table where customer is from delhi and older than 25
con.execute("SELECT*FROM Customers WHERE city='DELHI' AND age>=25;").fetchall()
```

```python
#Q9 Fetch orders where orders is either shoes or bag
con.execute("SELECT*FROM Orders WHERE product='Shoes' OR product='Bag';").fetchall()
```

```python
#Q10 Retrive orders where orice range in between 1000 to 2000
con.execute("SELECT*FROM Orders WHERE price  BETWEEN 1000 AND 2000;").fetchall()
```

```python
#Q11 Sort order by price in ascending orders
con.execute("SELECT*FROM Orders ORDER BY price ASC;").fetchall()
```

```python
#Q12 Group oders by customers and caluclte totla spending
con.execute("SELECT customer_id , SUM(price) FROM Orders GROUP BY customer_id;").fetchall()
```

```python
#Q13 Counts how many cusotmers ar from mumbai:
con.execute("SELECT COUNT(*) FROM Customers WHERE city='MUMBAI';").fetchall()
```

```python
#Q14 Filters grouped where to show customers spending above 5000
con.execute("SELECT customer_id , SUM(price) FROM Orders GROUP BY customer_id HAVING SUM(price)>5000;").fetchall()
```

```python
#Q15 Insert a new customer record in cusotmer table
con.execute("INSERT INTO Customers VALUES(11,'DHARMESH','JAIPUR',19);").fetchall()
con.execute("SELECT*FROM Customers WHERE customer_id=11;").fetchall()
```

```python
#Q16 Udate city of customer/-id ==1 to mumbai
con.execute("UPDATE Customers SET city='MUMBAI' WHERE customer_id=1;").fetchall()
con.execute("SELECT*FROM Customers;").fetchall()
```
