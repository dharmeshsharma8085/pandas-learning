Subsettting  data frames bases on some condition for a some row

```python
matrix_data= np.matrix("21,182,73;24,175,66;17,168,59;21,192,93;,24,185,86")
matrix_data
```

```python
r = ["A","B","C","D","E"]
c= ["Age","Height","Weight"]
df = pd.DataFrame(data = matrix_data, index = r , columns=c)
df
```

```python
print("\nRow with Height >180cm", "-"*35, sep ="")
df[df["Height"]>180]
```

```python
print("\nRows with height >180 and weight >80  ", "-"*30 , sep="")
df[(df["Height"]>180) & (df["Weight"]>80)]
```
