Introduction &   CREATION OF DATAFRAMSES

```python
import numpy as np
import pandas as pd
```

1 -: Using dict

```python
data = {
    "name" : ["a" , "b"  ,"c"],
    "age" : [12,12,13],
    "city" : ["Mumbai" , "Kolokata" , "Delhi"]
}

df = pd.DataFrame(data)
df
```

2 -: Using list of list

```python
data = [
     ["a" , "b"  ,"c"],
     [12,12,13],
    ["Mumbai" , "Kolokata" , "Delhi"]
]

df = pd.DataFrame(data , columns=["Name" , "Age" ,"City"])
df
```

3 -: From Numpy array

```python
arr = np.array([
    ["Amit" , 23 , "Mumbai"],
    ["Rahul " , 22 , "Hyd"],
    ["Ketan" , 30 , "Delhi"]
])

name=pd.Series(["Amit" , "Rahul" , "Ketan"])
age=pd.Series([23,22,30])
city=pd.Series(["Mumbai" , "Hyd" , "Delhi"])

df = pd.DataFrame({ "Name":name , "Age":age , "City":city})
df
```
