```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# The Iris dataset has no header, so specify column names
column_names = [
    'sepal_length',
    'sepal_width',
    'petal_length',
    'petal_width',
    'class'
]

# Read CSV without index column
df = pd.read_csv(
    url,
    header=None,
    names=column_names,
    index_col=False
)

print(df.head()) # head sai top5 millega
```
