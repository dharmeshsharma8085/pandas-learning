Multiindexing & Pivot Talbes in pandas

```python
arrays = [
    ["North" , "North" , "South" , "South"],
    [2024 , 2025 , 2024 , 2025]
]
```

```python
# Create multi index
index = pd.MultiIndex.from_arrays( arrays , names=("Region" ,"Year"))
index
```

```python
type(index)
```

```python
# cretaing DataFrame
df  = pd.DataFrame(
    np.random.randint(100,500, size = (4,2)),
    index= index,
    columns = ["Product A" , "Product B"]
)
df
```

```python
# print dataframe
df
```

```python
df.index
```

```python
df.columns
```

```python
# Slicing region
print(df.loc["North"])
print("--"*50)
print(df.loc["North" , 2024])
print("--"*50)
print(df.loc["North", 2025])
```

```python
# Slicing region
print(df.loc["South"])
print("--"*50)
print(df.loc["South" , 2024])
print("--"*50)
print(df.loc["South", 2025])
```

```python
df.xs("North" , level = "Region")  # another method for slicing in multidimensional
```

```python
#All region year 2024
print(df.loc[pd.IndexSlice[:,2024,:]])
```

```python
df.xs(2024 , level="Year")
```

```python
# Slice Prodouct
df.xs("Product A" , axis = 1)
# df.loc[: , "Product A"]
```

```python
df.xs("Product B", axis = 1)
```

```python
# Acess Specfic Year of a region ex south 2025
print(df.loc[("South", 2025)])
```

```python
df.xs(("South", 2025))
```

Pivot Table

```python
data = {
    "Region":["North","North","South","South"],
    "Year":[2024,2025,2024,2025],
    "Product":["PA","PB","PA","PB"],
    "Sales":[232 , 192 , 412 , 156]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
# Create a Pivot table : total sales per Region per Prodouct
pivot = pd.pivot_table( 
                       df,
                       values ="Sales",
                       index = "Region",
                       columns = "Product",
                       aggfunc = "sum")
pivot
```

Cross - tabulation

default output : counts/Frequencies

```python
data = {
    "Gender":["Male","Male","Female","Female","Male","Female","Male","Female"],
    "Preference":["Tea","Coffee","Coffee","Tea","Tea","Coffee","Coffee","Tea"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Create Cross Table using pivot table
pivot = df.pivot_table(
    df,
    index  = "Gender",
    columns = "Preference",
    aggfunc = "size"
    
)
pivot
```

```python
# Create Crosstab
pd.crosstab(df["Gender"], df["Preference"])
```
